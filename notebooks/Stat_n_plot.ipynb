### Descriptive statisitics

```python
import pandas as pd
import numpy as np
from scipy import stats

#Load csv to dataframes. List comprehension reads each file into a DataFrame, resulting in a list called dataframes
# Replace these with your actual file paths
filepaths = ["r0_rmsd.csv", "r1_rmsd.csv", "r2_rmsd.csv"]
dataframes = [pd.read_csv(path) for path in filepaths]

# Extract the metric of interest from each DataFrame
#This collects the mean RMSD per run into a list run_means.
variable = "nm_RMSD"
run_means = [df[variable].mean() for df in dataframes]

per_run_sd = [df[variable].std(ddof=1) for df in dataframes]
per_run_n = [df[variable].count() for df in dataframes]
per_run_sem = [sd / np.sqrt(n) for sd, n in zip(per_run_sd, per_run_n)]

print("Per-run statistics:")
for idx, (mean, sd, sem) in enumerate(zip(run_means, per_run_sd, per_run_sem), 1):
    print(f"Run {idx}: Mean = {mean:.4f}, SD = {sd:.4f}, SEM = {sem:.4f}")


# Overall mean and standard deviation across runs
mu = np.mean(run_means)
sigma = np.std(run_means, ddof=1)  # sample SD
n = len(run_means)
sem = sigma / np.sqrt(n)

# t-statistic for 95% CI, df = n-1
confidence = 0.95
t_value = stats.t.ppf((1 + confidence) / 2, df=n-1)
margin = t_value * sem
ci_lower = mu - margin
ci_upper = mu + margin

# Coefficient of Variation (percentage)
cv = (sigma / mu) * 100

#print results
print(f"Run means: {[round(x,4) for x in run_means]}")
print(f"Mean = {mu:.4f}")
print(f"Standard deviation = {sigma:.4f}")
print(f"Standard error = {sem:.4f}")
print(f"95% CI: [{ci_lower:.4f}, {ci_upper:.4f}] (t={t_value:.3f})")
print(f"CV = {cv:.2f}%")
```

### Mann-Whitney U and Kolmogorov-Smirnov Tests

```python
import pandas as pd
from scipy.stats import ks_2samp, mannwhitneyu

# load csv files
filepaths = ["r0_rmsd.csv", "r1_rmsd.csv", "r2_rmsd.csv"]  # Replace with your filenames
dataframes = [pd.read_csv(fp) for fp in filepaths]

# identify column name
variable = "nm_RMSD"  # Change as needed
columns = [df[variable] for df in dataframes]

# pairwise looping through statistical tests
from itertools import combinations

print(f"\nK-S and Mann-Whitney U tests for {variable}:")
for (i, data1), (j, data2) in combinations(enumerate(columns,1), 2):
    # Kolmogorov–Smirnov Test
    ks_stat, ks_p = ks_2samp(data1, data2)
    # Mann–Whitney U Test
    mw_stat, mw_p = mannwhitneyu(data1, data2, alternative='two-sided')
    print(f"Run {i} vs Run {j}:")
    print(f"  KS statistic = {ks_stat:.4f}, p = {ks_p:.4g}")
    print(f"  Mann-Whitney U = {mw_stat:.4f}, p = {mw_p:.4g}\n")
```

### ADF Test

```python
# Import required libraries
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Load your data
# Change 'your_file.csv' and column names as appropriate
df = pd.read_csv('ergo4ey7_rmsdBack.csv')

# If your time series is indexed by date, parse dates
# and set the time column as index (optional but recommended for series data)
df['Col1'] = pd.to_datetime(df['Col1'])
df.set_index('Col1', inplace=True)

# Replace 'value_column' with your target column name
time_series = df['Col2']

# Run ADF test
adf_result = adfuller(time_series)

# Print results
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')
for key, value in adf_result[4].items():
    print(f'Critical Value ({key}): {value}')

if adf_result[1] < 0.05:
    print('Result: The time series is stationary (reject null hypothesis).')
else:
    print('Result: The time series is non-stationary (fail to reject null hypothesis).')
```

### ANOVA (Log)

```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway

# Enumerate the CSV files to analyze, modify with your actual file paths
csv_files = [
    'r0_rmsd.csv',
    'r1_rmsd.csv',
    'r2_rmsd.csv'
    # Add more files as needed
]

column_of_interest = 'nm_RMSD'  # The column to log-transform and analyze

# Store log-transformed samples in a list
samples = []

for file in csv_files:
    df = pd.read_csv(file)
    # Ensure only positive, non-zero values for log transformation
    values = df[column_of_interest]
    log_values = np.log(values.replace(0, np.nan).dropna())
    samples.append(log_values.values)

# Perform one-way ANOVA on log-transformed samples
anova_result = f_oneway(*samples)

print(f'F-statistic: {anova_result.statistic}')
print(f'p-value: {anova_result.pvalue}')

if anova_result.pvalue < 0.05:
    print('Result: Significant difference between log(RMSD) means of the samples (reject null hypothesis).')
else:
    print('Result: No significant difference between log(RMSD) means of the samples (fail to reject null hypothesis).')
```

## MMPBSA / MMGBSA analyses

```python
# Jupyter Notebook Cell
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Input the filenames individually, not as folder
filenames = {
    'ergosterol': 'ergo_GBPBdelta.csv',
    'galantamine': 'galan_GBPBdelta.csv'
}

# Store results
results = {
    'Sample': [],
    'GB_mean': [],
    'GB_std': [],
    'GB_sem': [],
    'PB_mean': [],
    'PB_std': [],
    'PB_sem': []
}

for label, fname in filenames.items():
    df = pd.read_csv(fname)
    gb = df['GB_delta'].dropna().astype(float)
    pb = df['PB_delta'].dropna().astype(float)
    results['Sample'].append(label)
    results['GB_mean'].append(gb.mean())
    results['GB_std'].append(gb.std(ddof=1))
    results['GB_sem'].append(gb.std(ddof=1)/np.sqrt(len(gb)))
    results['PB_mean'].append(pb.mean())
    results['PB_std'].append(pb.std(ddof=1))
    results['PB_sem'].append(pb.std(ddof=1)/np.sqrt(len(pb)))

# Convert to DataFrame
stats_df = pd.DataFrame(results)

# Bar graph with error bars
fig, ax = plt.subplots(figsize=(8,5))

bar_width = 0.35
index = np.arange(len(stats_df))

# Plot GB values
gb_bar = ax.bar(index, stats_df['GB_mean'], bar_width, yerr=stats_df['GB_sem'],
                label='GB_delta', color='deepskyblue', capsize=5)
# Plot PB values next to GB
pb_bar = ax.bar(index + bar_width, stats_df['PB_mean'], bar_width, yerr=stats_df['PB_sem'],
                label='PB_delta', color='orange', capsize=5)

ax.set_xlabel('Sample')
ax.set_ylabel('Binding Energy (kcal/mol)')
ax.set_title('Binding Energies (mean ± SEM) for GB and PB methods')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(stats_df['Sample'])
ax.legend()

plt.tight_layout()
plt.savefig("ACHE_MM_bar.png", dpi=300)
plt.show()

# Pretty-print with 4 decimal places
for key, values in results.items():
    print(f"{key}:")
    formatted = [f"{v:.4f}" if isinstance(v, (int, float)) else str(v) for v in values]
    print("  " + ", ".join(formatted))
```

```python
import pandas as pd
from scipy.stats import mannwhitneyu

# Input files
samples = {
    'run 1': 'r0_MM.csv',
    'run 2': 'r1_MM.csv',
    'run 3': 'r2_MM.csv',
    'ctrl': 'Nctrl_MM.csv'
}

# Load data
dfs = {label: pd.read_csv(fname) for label, fname in samples.items()}

# Tests: controls vs each run
for run in ['run 1', 'run 2', 'run 3']:
    print(f"\n=== {run} vs ctrl ===")
    for col in ['GB_delta', 'PB_delta']:
        data_run = dfs[run][col].dropna()
        data_ctrl = dfs['ctrl'][col].dropna()
        # Mann-Whitney U test
        stat, p = mannwhitneyu(data_run, data_ctrl, alternative='two-sided')
        print(f"  {col}: U={stat:.2f}, p={p:.5e}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define input files (MMGBSA, each one contains GB_delta and PB_delta columns)
input_files = {
    'run 1': 'r0_MM.csv',
    'run 2': 'r1_MM.csv',
    'run 3': 'r2_MM.csv',
    'ctrl': 'Nctrl_MM.csv'
}

# Gather PB_delta data for all samples in one DataFrame
pb_delta_data = []
gb_delta_data = []
for label, fname in input_files.items():
    df = pd.read_csv(fname)
    pb = df['PB_delta'].dropna()
    gb = df['GB_delta'].dropna()
    pb_delta_data.append(pd.DataFrame({'Sample': label, 'PB_delta': pb}))
    gb_delta_data.append(pd.DataFrame({'Sample': label, 'GB_delta': gb}))

all_pb = pd.concat(pb_delta_data, ignore_index=True)
all_gb = pd.concat(gb_delta_data, ignore_index=True)

# Create violin plot for PB_delta
plt.figure(figsize=(8,6))
sns.violinplot(x='Sample', y='PB_delta', data=all_pb, inner='quartile', palette='Set2')
plt.title('Violin Plot of PB_delta Across Samples')
plt.ylabel('PB_delta (kcal/mol)')
plt.xlabel('Sample')
plt.tight_layout()
plt.show()

# Create violin plot for GB_delta
plt.figure(figsize=(8,6))
sns.violinplot(x='Sample', y='GB_delta', data=all_gb, inner='quartile', palette='Set1')
plt.title('Violin Plot of GB_delta Across Samples')
plt.ylabel('GB_delta (kcal/mol)')
plt.xlabel('Sample')
plt.tight_layout()
plt.show()
```
